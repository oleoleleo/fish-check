```python
import os
import glob
import random
import shutil
from icrawler.builtin import GoogleImageCrawler
```

```python
root_dir = '/Users/leotakahashi/workplace/fish_app/'
fishnames = ['アイゴ', 'オニカサゴ', 'カサゴ', 'カワハギ', 'キュウセンベラ', 'クサフグ', 'ソウシハギ', 'マハゼ', 'マアジ', 'マイワシ', 'ミノカサゴ', 'メジナ', 'メバル']
data_count=150
```

```python
def crawl_image(fishname, datacount, root_dir):
    
    crawler=GoogleImageCrawler(storage={'root_dir' : root_dir + fishname + '/train'})
    
    filters = dict(size = 'large', type = 'photo')
    
    crawler.crawl(keyword=fishname, filters=filters, max_num=datacount)
    
    if os.path.isdir(root_dir + fishname + '/test'):
        shutil.rmtree(root_dir + fishname + '/test')
    os.makedirs(root_dir + fishname + '/test')
    
    filelist=glob.glob(root_dir + fishname + '/train/*')
    
    test_ratio=0.2
    testfiles=random.sample(filelist, int(len(filelist) * test_ratio))
    
    for testfile in testfiles:
        shutil.move(testfile, root_dir + fishname + '/test/')
```

```python
for fishname in fishnames:
    crawl_image(fishname, data_count, root_dir)
```

```python
import glob
from azure.cognitiveservices.vision.customvision.prediction import CustomVisionPredictionClient
```

```python
base_url ='https://fish-teacher.cognitiveservices.azure.com/'
projectID ='0e061463-50be-41f6-8250-9d19bda4dda4'
publish_iteration_name='Iteration6'
prediction_key='de548dcab28148d6bc555e48663b2bda'
```

```python
def testModel(testfiles, fishname):
    data_count=len(testfiles)
    true_count=0
    
    predictor=CustomVisionPredictionClient(prediction_key, endpoint=base_url)
    
    for testfile in testfiles:
        predicts = {}
        with open(testfile, mode='rb') as f:
            results = predictor.classify_image(projectID, publish_iteration_name, f.read())
            
            for prediction in results.predictions:
                predicts[prediction.tag_name] = prediction.probability
            
            prediction_result= max(predicts, key= predicts.get)
            
            if fishname == prediction_result:
                true_count += 1
        
    accuracy= (true_count / data_count) * 100
    print('正解率：' + str(accuracy) + '%')
```

```python
root_dir = '/Users/leotakahashi/workplace/fish_app/'
fishnames = ['ゴンズイ', 'イシガキダイ']


for fishname in fishnames:
    print('****' + fishname + '****')
    testfiles = glob.glob(root_dir + fishname + '/test/*')
    testModel(testfiles, fishname)
```
